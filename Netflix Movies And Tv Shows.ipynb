#### Chart - 8   Line Plot — Number of Titles Added Per Year


````python
# Chart - 8 visualization code

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure 'date_added' is in datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract year and type
df['year_added'] = df['date_added'].dt.year
df_filtered = df.dropna(subset=['year_added'])

# Group by year and type
type_by_year = df_filtered.groupby(['year_added', 'type']).size().unstack().fillna(0)

# Plot
plt.figure(figsize=(12, 6))
type_by_year.plot(kind='bar', stacked=True, colormap='Accent', figsize=(12, 6))
plt.title("Movies vs. TV Shows Added to Netflix Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.legend(title="Type")
plt.tight_layout()
plt.show()
````

#### Chart - 9   Heatmap — Monthly Content Additions by Year

````python
# Chart - 9 visualization code

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure correct datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract year and month
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

# Create pivot table
monthly_trend = df.pivot_table(index='month_added', columns='year_added', 
                                values='show_id', aggfunc='count')

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(monthly_trend, cmap="YlGnBu", linewidths=.5, linecolor='gray', annot=True, fmt=".0f")
plt.title("Monthly Netflix Title Additions by Year")
plt.xlabel("Year")
plt.ylabel("Month")
plt.yticks(ticks=range(1,13), labels=[
    "Jan", "Feb", "Mar", "Apr", "May", "Jun", 
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"], rotation=0)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

Answer Here.

The heatmap was selected to reveal patterns over two time dimensions — month and year. It allows a quick visual interpretation of when content is most frequently released, which would be hard to analyze with a traditional line or bar plot.
